```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras
from keras.utils import load_img, img_to_array, array_to_img
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = load_img('/content/cat.jpg', target_size = (200,200))
```

```python
img
# plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 40,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    vertical_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = 'reflect'
)
```

```python
img_arr = img_to_array(img)
```

```python
img_arr.shape
```

```python
input_batch = img_arr.reshape(1,200,200,3)
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size = 1, save_to_dir = 'cat2'):
  i = i+1
  if i==20:
    break
```
